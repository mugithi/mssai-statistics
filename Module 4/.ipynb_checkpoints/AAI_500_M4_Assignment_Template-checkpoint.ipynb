## Problem # 4.2.

For a sequence of observations of a binary random variable, you observe the geometric random
variable (Section 2.2.2) outcome of the first success on observation number $y = 3$. Find and plot
the likelihood function.

`our answer goes here`

Geometric Probability distribution for a binary outcome

Probability of success = $\pi$

Probability of = $(1-\pi)$

For success to occur at $y=3$

$f(y) = (1-\pi)^{y-1} \pi$


$l(\pi)=(1 - \pi) (1-\pi)\pi$

$l(\pi) = (1 - \pi)^2  \pi$


```python
import numpy as np
import matplotlib.pyplot as plt

p = np.linspace(0, 1, 100) # Define the range of p values
likelihood = (1-p)**2 * p # Calculate the likelihood function

plt.plot(p, likelihood) # Plot the likelihood function
plt.xlabel('p')
plt.ylabel('Likelihood')
plt.title('Likelihood function for y = 3')
plt.show()
```

## Problem # 4.4.

For the `Students` data file (Exercise 1.2 in Chapter 1) and corresponding population, find the
ML estimate of the population proportion believing in life after death. Construct a Wald 95%
confidence interval, using its formula (4.8). Interpret.

`Your answer goes here`

```python
# Read in the Students data file
students = pd.read_csv('https://stat4ds.rwth-aachen.de/data/Students.dat', sep='\s+')

# View the first few 10 rows of the dataset
students['life'].notnull().sum()
```

$\hat{\pi} = \frac{(Y)\space number\space of\space successes\space in\space sample}{n}$

```python
# life = belief in life after death (1 = yes, 2 = no, 3 = undecided)
# Count all students who beleive 1
life = students[students['life'] == 3].shape[0]
total = students['life'].notnull().sum()
point_estimate = life/total 
print("ML Estimate of population propotion", round(point_estimate,2)) 
```

```python
#Compute the point estimate of the parameter based on the sample data for the sample proportion.

life = students[students['life'] == 3].shape[0]
total = students['life'].notnull().sum()
pi_hat = life/total 
print("Point estimate of population propotion", round(pi_hat,2)) 

#Calculate standard error of the sample proportion
standard_error = math.sqrt(pi_hat * (1 - pi_hat) / total)

print("Standard error: ", round(standard_error,2))

#Find the z_apha that corresponds to a 95% confidence level.

z_apha = stats.norm.ppf((1+.95)/2)
print("Z_apha",round(z_apha,2))

#Compute the lower and upper bounds of the confidence interval

lower_bound = pi_hat - (z_apha * standard_error)
upper_bound = pi_hat + (z_apha * standard_error)


print("We are 95% confident that between", round(lower_bound,2)*100, "% and", round(upper_bound,2)*100,"% of the population belive in life after death")
```

## Problem # 4.11.

The observations on number of hours of daily TV watching for the 10 subjects in the 2018 GSS
who identified themselves as Islamic were 0, 0, 1, 1, 1, 2, 2, 3, 3, 4.

(a)  Construct and interpret a 95% confidence interval for the population mean.
(b)  Suppose the observation of 4 was incorrectly recorded as 24. What would you obtain for
the 95% confidence interval? What does this suggest about potential effects of outliers on
confidence intervals for means?

`(a) Your answer goes here`

```python
# Create an array 
tv_watchers = np.array([0, 0, 1, 1, 1, 2, 2, 3, 3, 4])

def conf_level(array):
    # Calculate the population mean
    mu = array.mean(); print("Mean:", mu)

    #Calculate the standard error 
    std = array.std(); print("Standard Deviation:", round(std,2))#standard deviation 
    std_error = std/np.sqrt(len(array)); print("Standard Error:", round(std_error,2)) #standard error 

    #Calculate the critical value for 95% 
    z_alpha = stats.norm.ppf((1+.95)/2)

    #Calculate the EBM 
    ebm = z_alpha*std_error

    #Calculate the  upper bound and lower bound mean 
    upper_bound = mu + ebm; print("Upper Bound:", round(upper_bound,2))
    lower_bound = mu - ebm; print("Lower Bound:", round(lower_bound,2))

    print("We are 95% confident that subjects who identified themselves as Islamic watched between",round(lower_bound,2),"hrs and ", round(upper_bound,2), "hours of tv in the 2018 GSS study")
    
conf_level(tv_watchers)
```

`(b) Your answer goes here`

```python
tv_watchers_outlier = np.array([0, 0, 1, 1, 1, 2, 2, 3, 3, 24])
conf_level(tv_watchers_outlier)
```

With the outlier, it increaes the standard deviation which increases the standard error. As a result the Wald confidence interval has a negative value for the lower bound

```python
def conf_level(array1, array2, name1, name2, confidence_decimal):
    # Calculate the population mean
    mu1 = array1.mean(); print(name1, "Mean:", round(mu1,2))
    mu2 = array2.mean(); print(name2, "Mean:", round(mu2,2))

    #Calculate the standard error 
    std1 = array1.std(); print(name1, "Standard Deviation:", round(std1,2))#standard deviation 
    std_error1 = std1/np.sqrt(len(array1)); print(name1, "Standard Error:", round(std_error1,2)) #standard error 
    
    #Calculate the standard error 
    std2 = array2.std(); print(name2, "Standard Deviation2:", round(std2,2))#standard deviation 
    std_error2 = std2/np.sqrt(len(array2)); print(name2, "Standard Error:", round(std_error2,2)) #standard error 

    #Calculate the critical value for 95% 
    z_alpha = stats.norm.ppf((1+confidence_decimal)/2)

    #Calculate the EBM 
    ebm1 = z_alpha*std_error1
    ebm2 = z_alpha*std_error2

    #Calculate the  upper bound and lower bound mean 
    upper_bound1 = mu1 + ebm1 # print(name1, "Upper Bound1:", round(upper_bound1,2))
    lower_bound1 = mu1 - ebm1 # print(name1, "Lower Bound1:", round(lower_bound1,2))
    
    #Calculate the  upper bound and lower bound mean 
    upper_bound2 = mu2 + ebm2 # print(name2, "Upper Bound2:", round(upper_bound2,2))
    lower_bound2 = mu2 - ebm2 # print(name2, "Lower Bound2:", round(lower_bound2,2))
    
    
    print("We are",round(confidence_decimal*100,2),"% confident that",name1,"watch between",round(lower_bound1,2),"hrs and", round(upper_bound1,2), "hours of tv a week")
    print("We are",round(confidence_decimal*100,2),"% confident that",name2,"watch between",round(lower_bound2,2),"hrs and", round(upper_bound2,2), "hours of tv a week")
    

    #difference between the population proportions, we can use the formula
    difference_upper_bound = (mu1-mu2)+(z_alpha)*np.sqrt(np.float_power(std_error1,2)+np.float_power(std_error2,2))#; print(difference_upper_bound)
    difference_lower_bound = (mu1-mu2)-(z_alpha)*np.sqrt(np.float_power(std_error1,2)+np.float_power(std_error2,2))#; print(difference_lower_bound)
    
    print("We are",round(confidence_decimal*100,2),"% confident that the difference between", name1,"and", name2, "is between", round(difference_lower_bound,2), "and", round(difference_upper_bound,2), "a week")

#Students
students = pd.read_csv('https://stat4ds.rwth-aachen.de/data/Students.dat', sep='\s+')
confidence_level = .95


#Male Students
male = "Male Students"
male_students = students[students["gender"] == 0].drop(columns=["subject", "gender", "age", "hsgpa" , "cogpa", "dhome", "dres", "sport", "news" , "aids", "veg", "affil", "ideol", "relig", "abor", "affirm", "life"]).squeeze()


#Female Students
female = "Female Students"
female_students = students[students["gender"] == 1].drop(columns=["subject", "gender", "age", "hsgpa" , "cogpa", "dhome", "dres", "sport", "news" , "aids", "veg", "affil", "ideol", "relig", "abor", "affirm", "life"]).squeeze()

conf_level(male_students, female_students, male, female, confidence_level)
```

## Problem # 4.15.

In the 2018 General Social Survey, when asked whether they believed in life after death, 1017
of 1178 females said yes, and 703 of 945 males said *yes*. Construct 95% confidence intervals
for the population proportions of females and males that believe in life after death and for the
difference between them. Interpret.

`Your answer goes here`

```python
# Calculate the population propotion
confidence_decimal = 0.95
#Males
name1 ="Females"
name2 ="Males"
n1 = 1178 # Sample Size
mu1 = 1017/n1 # Sample Propotion 

#Females
n2 = 1178 # Sample Size
mu2 = 703/n2 #Sample Propotion 

#Calculate the standard error Males
std_error1 = np.sqrt(mu1*(1-mu1)/n1)


#Calculate the standard error Females
std_error2 = np.sqrt(mu2*(1-mu2)/n2)

#Calculate the critical value for 95% 
z_alpha = stats.norm.ppf((1+confidence_decimal)/2)

#Calculate the EBM 
ebm1 = z_alpha*std_error1
ebm2 = z_alpha*std_error2

#Calculate the  upper bound and lower bound mean 
upper_bound1 = mu1 + ebm1 # print(name1, "Upper Bound1:", round(upper_bound1,2))
lower_bound1 = mu1 - ebm1 # print(name1, "Lower Bound1:", round(lower_bound1,2))

#Calculate the  upper bound and lower bound mean 
upper_bound2 = mu2 + ebm2 # print(name2, "Upper Bound2:", round(upper_bound2,2))
lower_bound2 = mu2 - ebm2 # print(name2, "Lower Bound2:", round(lower_bound2,2))


print("We are",round(confidence_decimal*100,2),"% confident that between",round(lower_bound1*100,2),"% and", round(upper_bound1*100,2),"of",name1, "believe in life after death")
print("We are",round(confidence_decimal*100,2),"% confident that between",round(lower_bound2*100,2),"% and", round(upper_bound2*100,2),"of",name2, "believe in life after death")


#difference between the population proportions, we can use the formula
difference_upper_bound = (mu1-mu2)+(z_alpha)*np.sqrt(np.float_power(std_error1,2)+np.float_power(std_error2,2))#; print(difference_upper_bound)
difference_lower_bound = (mu1-mu2)-(z_alpha)*np.sqrt(np.float_power(std_error1,2)+np.float_power(std_error2,2))#; print(difference_lower_bound)

print("We are",round(confidence_decimal*100,2),"% confident that the difference between the", name1,"and", name2, "who believe in life after death is between", round(difference_lower_bound*100,2), "% and", round(difference_upper_bound*100,2), "%")

```

```python
# We use standard normal distribution with a mean of 0 and std of 1, We then use a 
# Cumulative distribution function (CDF) of a standard normal distribution. The CDF at a given point y
# will give us the probability that a standard normal random variable will take a value less than or equal to y

#P (Y) = 1
#P (Y > 3) = 1 - CDF(3)
mean = 0
std = 1
y = 3
P_Y_gt_3_norm = 1 - norm.cdf(y, mean, std)
print("The Probability of Y, taking a value greater than 3 is small at ",round(P_Y_gt_3_norm*100,2),"%")
```
